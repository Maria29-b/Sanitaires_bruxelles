## 4. Questions pour l'analyse croisée

En me basant uniquement sur l'exploration visuelle des données, voici les trois questions auxquelles je souhaite répondre grâce à l'analyse croisée :

### Question 1 :
**Comment les toilettes publiques et les urinoirs sont-ils répartis géographiquement par quartier ?**
- Objectif : Identifier les quartiers avec la meilleure et la pire couverture en installations sanitaires
- Croisement : Regrouper les deux datasets par quartier (Quartier/Wijk) et compter les occurrences

### Question 2 :
**Quelle est l'accessibilité PMR (Personnes à Mobilité Réduite) des toilettes publiques par rapport aux urinoirs ?**
- Objectif : Évaluer si les urinoirs offrent des équipements accessibles et comparer avec les toilettes
- Croisement : Analyser les colonnes Accessibilité/Toegankelijkheid/Accessibility

### Question 3 :
**Les zones (proches) des urinoirs disposent-elles de toilettes publiques à proximité ?**
- Objectif : Vérifier si les hommes peuvent facilement trouver des toilettes complètes à côté des urinoirs
- Croisement : Utiliser les coordonnées Geo Point pour calculer les distances et identifier les zones sous-équilibrées
